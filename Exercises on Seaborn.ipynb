```python
import seaborn as sns
%matplotlib inline
```

```python
sns.get_dataset_names() #Some built in datasets in Seaborn
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

# Types of Distribution Plot:

```python
#distplot is replaced by displot in future versions and 'kde' stands for 'Kernel Density Estimation'
#Univariate distribution
sns.displot(tips['total_bill'],kde='true',bins=5) 
```

# Joint Plot:

```python
#Bivariate distribution
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

# Pair Lot:

```python
#Pair Plot distribution which shows the pairwise relationship across the entire dataframe
sns.pairplot(tips,hue='sex')
```

# RugPlot:

```python
#Plotting by drawing ticks 
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# Types of Categorical data plots:

# Bar Plots:

```python
#Estimator is used to estimate mathematically, std refers to 'standard deviation'

import numpy as np
sns.barplot(data=tips,y='total_bill',x='sex',estimator=np.std)
```

# Count Plot:

```python
sns.countplot(data=tips,x='sex')
```

# Box Plot:

```python
#bivariate which has one numerical and one categorical variable
sns.boxplot(data=tips,y='total_bill',x='day',hue='smoker')
```

```python
#Orientation parameter which exhibits the box plot horizontally / vertically by using 'h' or 'v'
sns.boxplot(data=tips,orient='h')
```

# Violin Plot:

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
#Splitting the plot itself by 'sex' attribute
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

# Strip Plot:

```python
#Jitter is set to 'true' by default and 'split' is replaced by 'dodge'
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
```

# Swarm Plot:

```python
sns.swarmplot(data=tips,x='day',y='total_bill')
```

# Combining the Categorical Plots:

```python
sns.violinplot(data=tips,x='day',y='total_bill')
sns.swarmplot(data=tips,x='day',y='total_bill',color='black')
```

# Factor Plot:

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='bar')
```

# Working on Grids:

PairGrid,FacetGrid and JointGrid:

```python
ir=sns.load_dataset('iris')
```

```python
import matplotlib.pyplot as plt
sns.PairGrid(ir)
```

```python
g=sns.PairGrid(ir)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(ir)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(ir,hue='species')
```

```python
#Regression Plot with respect to the sex and also the changed marker size
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['v','o'],scatter_kws={'s':100})
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='sex')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.displot(titanic['fare'],kind='hist',bins=25,color='red',palette='coolwarm')
```

```python
sns.boxplot(data=titanic,x='class',y='age',palette='rainbow')
```

```python
sns.swarmplot(data=titanic,x='class',y='age',size=5,palette='Set2')
```

```python
#Refer Matplotlib styles and color for various color combinations
sns.countplot(data=titanic,x='sex',palette='viridis')
```

```python
tcorr=titanic.corr()
tcorr
```

```python
sns.heatmap(tcorr,cmap='coolwarm')
plt.title("titanic.corr()")
```

```python
g=sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```
